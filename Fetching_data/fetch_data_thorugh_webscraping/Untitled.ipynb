```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://books.toscrape.com/catalogue/page-1.html'
headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
webpage=requests.get(url,headers=headers)
```

```python
import lxml
print(lxml.__version__)
```

```python
soup=BeautifulSoup(webpage.text,'lxml')
print(soup.title.text)
```

```python
for i in soup.find_all("h3"):
    print(i.text.strip())
```

```python
book_container=soup.find('ol',class_="row")
books = book_container.find_all('article', class_='product_pod')
len(books)
```

```python
name = []
rating=[]
price=[]
availabality=[]

for i in books:
    title = i.find('h3').a['title']
    name.append(title.strip())
    rating_tag=i.find('p',class_='star-rating')
    rating.append(rating_tag['class'][1])
    price.append(i.find('p',class_="price_color").text.strip())
    availabality.append(i.find('p', class_="instock availability").text.strip())

print(availabality)
```

```python
d={"book_name":name,"rating":rating ,"price":price,"availabality":availabality}
df=pd.DataFrame(d)
final=pd.DataFrame(df)
final
```

```python
name = []   
rating=[]
price=[]
availabality=[]
for j in range(1,51):
    url = f'https://books.toscrape.com/catalogue/page-{j}.html'
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
    webpage=requests.get(url,headers=headers)
    soup=BeautifulSoup(webpage.text,'lxml')
    book_container=soup.find('ol',class_="row")
    books = book_container.find_all('article', class_='product_pod')
   

    for i in books:
       title = i.find('h3').a['title']
       name.append(title.strip())
       rating_tag=i.find('p',class_='star-rating')
       rating.append(rating_tag['class'][1])
       price.append(i.find('p',class_="price_color").text.strip())
       availabality.append(i.find('p', class_="instock availability").text.strip())

d={"book_name":name,"rating":rating ,"price":price,"availabality":availabality}
final=pd.DataFrame(d)
```
