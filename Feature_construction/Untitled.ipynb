## Apllying feature construction

```python
X['Family_size']=X['SibSp']+X['Parch']+1
```

```python
X
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
myfunc(1)
```

```python
X['Family_type']=X['Family_size'].apply(myfunc)
```

```python
X
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=14))
```
