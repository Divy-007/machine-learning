```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df
```

```python
#function for row sampling 
def sample_rows(df,percent):
  return df.sample(int(percent*df.shape[0]),replace=True)
```

```python
df1=sample_rows(df,0.5)
print(df1.shape)
df1.head()
```

```python
df2=sample_rows(df,0.5)
print(df2.shape)
df2.head()
```

```python
df3=sample_rows(df,0.5)
print(df3.shape)
df3.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1=DecisionTreeClassifier()
clf2=DecisionTreeClassifier()
clf3=DecisionTreeClassifier()
```

```python
clf1.fit(df1.iloc[:,0:5],df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:5],df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:5],df3.iloc[:,-1])
```

```python
from sklearn.tree import plot_tree
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```

```python
#function for feature sampling 
import random
def feature_sampling(df,percent):
  cols = random.sample(df.columns.tolist()[:-1],int(percent*(df.shape[1]-1)))# excluding target column
  new_df = df[cols]
  new_df['target'] = df['target']
  return new_df
```

```python
df1=feature_sampling(df,0.5)
df1.shape
```

```python
df2=feature_sampling(df,0.5)
```

```python
df3=feature_sampling(df,0.5)
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
     
```

```python
# combined sampling
def combined_sampling(df,row_percent,col_percent):
  new_df = sample_rows(df,row_percent)
  return feature_sampling(new_df,col_percent)
```

```python
df1 = combined_sampling(df,0.5,0.5)
```

```python
df2 = combined_sampling(df,0.5,0.5)
```

```python
df3 = combined_sampling(df,0.5,0.5)
```
