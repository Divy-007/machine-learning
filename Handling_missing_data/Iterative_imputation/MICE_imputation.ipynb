```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df.head()
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df.head()
```
