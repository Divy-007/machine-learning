```python
import instaloader
from instaloader import Profile, PrivateProfileNotFollowedException, LoginRequiredException

# 1. Create Instaloader instance
L = instaloader.Instaloader()

# 2. OPTIONAL: Log in (required for private profiles)
# ----------------------------------------------------
# L.login("your_username", "your_password")               # Direct (not recommended)
# L.interactive_login("your_username")                    # Safer, saves session
# L.load_session_from_file("your_username")               # Reuse saved session

# 3. Fetch the profile
TARGET_USERNAME = "sen.orita_09"  # Change as needed

try:
    profile = Profile.from_username(L.context, TARGET_USERNAME)
except Exception as e:
    print(f"Failed to load profile: {e}")
    exit(1)

# 4. Print basic profile metadata
print(f"Username      : {profile.username}")
print(f"Full Name     : {profile.full_name}")
print(f"Bio           : {profile.biography}")
print(f"Profile Pic   : {profile.profile_pic_url}")
print(f"Followers     : {profile.followers}")
print(f"Following     : {profile.followees}")
print(f"Posts Count   : {profile.mediacount}")
print(f"Is Private    : {profile.is_private}")
print(f"Is Verified   : {profile.is_verified}")
print(f"External URL  : {profile.external_url}")
print(f"User ID       : {profile.userid}")
print("-" * 50)

# 5. Iterate over posts (public or private IF you are followed)
print("Latest posts (shortcode, likes, caption preview):")

try:
    for i, post in enumerate(profile.get_posts()):
        # Limit to first 5 posts to avoid flooding
        if i >= 5:
            print("... (stopping after 5 posts)")
            break

        caption = (post.caption[:60] + "...") if post.caption and len(post.caption) > 60 else post.caption
        print(f"{i+1}. {post.shortcode} | Likes: {post.likes} | Caption: {caption or '(no caption)'}")

        # OPTIONAL: Download media + metadata
        # L.download_post(post, target=profile.username)

except PrivateProfileNotFollowedException:
    print("This profile is PRIVATE and does NOT follow you â†’ posts are inaccessible.")
except LoginRequiredException:
    print("Login required to view this profile's posts.")
except Exception as e:
    print(f"Error while fetching posts: {e}")
```

```python
import instaloader

L = instaloader.Instaloader()

# Try to load session silently
try:
    L.load_session_from_file("devxdivy")  # Replace with your IG username
    if L.context.is_logged_in:
        print("LOGGED IN")
        print(f"Username: {L.context.username}")
        print(f"User ID: {L.context.userid}")
    else:
        print("NOT logged in (session invalid)")
except FileNotFoundError:
    print("NOT logged in (no session file)")
except Exception as e:
    print(f"Error: {e}")
```
