```python
# Current EXACT class combo for username/full name spans on IG
spans = soup.find_all(
    "span",
    class_="x1lliihq x193iq5w x6ikm8r x10wlt62 xlyipyv xuxw1ft x1mh8g0r"
)

# Or the even more common variant (with one extra class)
spans = soup.find_all(
    "span",
    class_="x1lliihq x6ikm8r x10wlt62 xlyipyv xuxw1ft _acan _acao _acat _acaw _a6hd"
)

# Extract the visible text — NOT title! (IG moved names to .text)
names = [span.get_text(strip=True) for span in spans if span.get_text(strip=True)]

print("Found spans:", len(spans))
print("Names:", names)
first_name = names[0] if names else None
print("First:", first_name)
```

```python
def is_instagram_name_span(tag):
    if tag.name != "span":
        return False
    classes = tag.get("class", [])
    # Core IG name classes — these rarely change fully
    core = ["x1lliihq", "x6ikm8r", "x10wlt62", "xlyipyv", "xuxw1ft"]
    return sum(c in classes for c in core) >= 4

spans = soup.find_all(is_instagram_name_span)

names = []
for span in spans:
    text = span.get_text(strip=True)
    if text and len(text) > 1 and text not in ["·", "Follow", "Following"]:
        names.append(text)

print("Instagram names found:", list(dict.fromkeys(names)))  # dedupe
```
